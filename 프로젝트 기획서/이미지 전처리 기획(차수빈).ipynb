```python
# 파일의 이름을 재정의 해주는 함수 정의
def rename(files):

  if 'angry' in files[0]: # 파일 경로에 angry라는 단어가 포함되어 있으면
     for i,f in enumerate(files): # i: 순번, f: 파일
         os.rename(f, os.path.join(path+"/angry", 'angry_' + '{0:03d}.jpg'.format(i)))
     angry = glob.glob(path+"/angry" + '/*')    
     print("angry {}번째 이미지까지 성공".format(i+1))

  elif 'disgust' in files[0]:
     for i,f in enumerate(files):
         os.rename(f, os.path.join(path+"/disgust", 'disgust_' + '{0:03d}.jpg'.format(i)))
     disgust = glob.glob(path+"/disgust"+'/*')
     print("disgust {}번째 이미지까지 성공".format(i+1))

  elif 'fear' in files[0]:
     for i,f in enumerate(files):
         os.rename(f, os.path.join(path+"/fear", 'fear_' + '{0:03d}.jpg'.format(i)))
     fear = glob.glob(path+"/fear"+'/*')
     print("fear {}번째 이미지까지 성공".format(i+1))

  elif 'happy' in files[0]:
    for i,f in enumerate(files):
         os.rename(f, os.path.join(path+"/happy", 'happy_' + '{0:03d}.jpg'.format(i)))
    happy = glob.glob(path+"/happy"+'/*')
    print("happy {}번째 이미지까지 성공".format(i+1))

  elif 'neutral' in files[0]:
     for i,f in enumerate(files):
         os.rename(f, os.path.join(path+"/neutral", 'neutral_' + '{0:03d}.jpg'.format(i)))
     neutral = glob.glob(path+"/neutral"+'/*')
     print("neutral {}번째 이미지까지 성공".format(i+1))

  elif 'sad' in files[0]:
     for i,f in enumerate(files):
         os.rename(f, os.path.join(path+"/sad", 'sadr_' + '{0:03d}.jpg'.format(i)))
     sad = glob.glob(path+"/sad"+'/*')
     print("sad {}번째 이미지까지 성공".format(i+1))

  elif 'surprise' in files[0]:
    for i,f in enumerate(files):
      os.rename(f, os.path.join(path+"/surprise", 'surprise_' + '{0:03d}.jpg'.format(i)))
    surprise = glob.glob(path+"/surprise"+'/*')
    print("surprise {}번째 이미지까지 성공".format(i+1))
```

# **2. 이미지 시각화하기**
- 파일이 잘 불러지는지 테스트하기 위해 몇 개만 가져다가 matplotlib으로 찍어본다.

```python
# 필요한 모듈 import
import random # 일부 이미지만 랜덤으로 가져올 것이기 때문에
import matplotlib.pyplot as plt 
from matplotlib.pyplot import imshow
import cv2 # 시각화
%matplotlib inline
```

```python
# 이미지를 입맛에 맞게 변형하여 불러올 수 있도록 해주는 함수 정의하기
def get_img(file_path): 
  img_li = cv2.imread(file_path,cv2.IMREAD_GRAYSCALE) # 사진을 흑백으로 읽어오기(학습 시 배경의 영향 약화하기 위함)
  img_li = cv2.cvtColor(img_li, cv2.COLOR_BGR2RGB) # 사진 색상 변경해주기
  img_li = cv2.resize(img_li,28*28) # 일단 사진 크기 그냥 임의로 설정해 봄

  return img_li
```

```python
# 각각의 감정 이미지 폴더에서 2개의 사진씩 랜덤으로 뽑아와서 이미지 시각화해보기

# 이미지 리스트 준비하기

import random # 이미지 랜덤으로 뽑아올 거니까!

img_lists = [] # 랜덤으로 뽑아온 이미지들을 싹 다 저장(앞에서부터 각 감정별 2개의 사진이 저장될 예정)

img_num = range(0,6200) # 가장 개수가 많은 happy에 범위를 맞춤

for i in random.sample(img_num,2): # 각각 2개씩 뽑아오기

  img_lists.append(get_img(angry[i]))
  img_lists.append(get_img(disgust[i]))
  img_lists.append(get_img(fear[i]))
  img_lists.append(get_img(happy[i]))
  img_lists.append(get_img(neutral[i]))
  img_lists.append(get_img(sad[i]))
  img_lists.append(get_img(surprise[i]))

# 시각화를 할 그림판 설정하기(격자)
cols = ['angry','disgust','fear','happy','neutral','sad','surprise'] # 감정별로 naming

for i in range(14): # 2행 7열의 데이터
  plt.subplot(7,2,i+1) # 7행 2열로 그래프 개요 그려놓고 i+1번째에 위치시키기(1번부터 시작하기에 I+1이다.)
  plt.xticks([])
  plt.yticks([])
  plt.grid(False) # 격자 표시 x
  plt.imshow(img_lists[i],cmap = 'gray') # 배경: 흑백으로 날려버리기
  plt.title(cols[i//2]+'_'+str(i%2+1)) # 예시: 0번째 사진 => 'angry_1'로 네이밍

plt.show()
```

# **3. train_data/test_data splitting**

```python
path = '/content/drive/MyDrive/AI 감정인식 프로젝트/datasets'

angry = sorted(glob.glob(path+"/angry" + '/*'))
disgust = sorted(glob.glob(path+"/disgust"+'/*'))
fear = sorted(glob.glob(path+"/fear"+'/*'))
happy = sorted(glob.glob(path+"/happy"+'/*'))
neutral = sorted(glob.glob(path+"/neutral"+'/*'))
sad = sorted(glob.glob(path+"/sad"+'/*'))
surprise = sorted(glob.glob(path+"/surprise"+'/*'))

print('angry 이미지 개수: {}'.format(len(angry)))
print('disgust 이미지 개수: {}'.format(len(disgust)))
print('fear 이미지 개수: {}'.format(len(fear)))
print('happy 이미지 개수: {}'.format(len(happy)))
print('neutral 이미지 개수: {}'.format(len(neutral)))
print('sad 이미지 개수: {}'.format(len(sad)))
print('surprise 이미지 개수: {}'.format(len(surprise)))
```

```python
# 각각의 감정에 해당하는 dataset을 train과 test로 split
# 상대적으로 비율이 적은 test_data(0.2)를 분리해 내는 방식을 이용

import math

# 감정별로 전체 사진의 약 20%(반올림해서<- 사진을 반토막 낼 순 없으니..)의 사진을 test_data로 이용하겠다.
angry_test_cnt = round(len(angry)*0.2) 
disgust_test_cnt = round(len(disgust)*0.2)
fear_test_cnt = round(len(fear)*0.2)
happy_test_cnt = round(len(happy)*0.2)
neutral_test_cnt = round(len(neutral)*0.2)
sad_test_cnt = round(len(sad)*0.2)
surprise_test_cnt = round(len(surprise)*0.2)

print('angry_test 파일에 들어갈 이미지 개수: {}/{}'.format(angry_test_cnt,len(angry)))
print('disgust_test 파일에 들어갈 이미지 개수: {}/{}'.format(disgust_test_cnt,len(disgust)))
print('fear_test 파일에 들어갈 이미지 개수: {}/{}'.format(fear_test_cnt,len(fear)))
print('happy_test 파일에 들어갈 이미지 개수: {}/{}'.format(happy_test_cnt,len(happy)))
print('neutral_test 파일에 들어갈 이미지 개수: {}/{}'.format(neutral_test_cnt,len(neutral)))
print('sad_test 파일에 들어갈 이미지 개수: {}/{}'.format(sad_test_cnt,len(sad)))
print('surprise_test 파일에 들어갈 이미지 개수: {}/{}'.format(surprise_test_cnt,len(surprise)))
```

```python
import shutil 
# 파일을 복사해주는 파이썬 모듈
# shutil.copy(“대상 파일”, “복사본을 저장할 파일명”)
```

```python
# 파일을 원하는 경로에 복사해주는 함수 정의하기
def data_split(img_list, test_count, train_path, test_path): 
  
  test_files=[]
  for i in random.sample(img_list, test_count): # 20%에 해당하는 사진 개수만큼 랜덤으로 뽑아서 test_data로 사용
    test_files.append(i)

  # 차집합으로 train/test 리스트 생성하기
  # img_list에 있는 파일 중test_files가 아닌 모든 사진들을 train_files 리스트로 저장
  train_files = [x for x in img_list if x not in test_files] 

  for k in train_files:
    shutil.copy(k, train_path)
  
  for c in test_files:
    shutil.copy(c, test_path)

  print('train 폴더 이미지 개수 : {}\ntest 폴더 이미지 개수 : {}'.format(len(glob.glob(train_path+'/*')),len(glob.glob(test_path+'/*'))))
```

```python
angry_train_path='' # angry의 train_data가 저장될 폴더의 경로(미리 생성해 두기!)
angry_test_path='' # angry의 test_data가 저장될 폴더의 경로

disgust_train_path='' # disgust의 train_data가 저장될 폴더의 경로(미리 생성해 두기!)
disgust_test_path='' # disgust의 test_data가 저장될 폴더의 경로

fear_train_path='' # fear의 train_data가 저장될 폴더의 경로(미리 생성해 두기!)
fear_test_path='' # fear의 test_data가 저장될 폴더의 경로

happy_train_path='' # happy의 train_data가 저장될 폴더의 경로(미리 생성해 두기!)
happy_test_path='' # happy의 test_data가 저장될 폴더의 경로

neutral_train_path='' # neutral의 train_data가 저장될 폴더의 경로(미리 생성해 두기!)
neutral_test_path='' # neutral의 test_data가 저장될 폴더의 경로

sad_train_path='' # sad의 train_data가 저장될 폴더의 경로(미리 생성해 두기!)
sad_test_path='' # sad의 test_data가 저장될 폴더의 경로

surprise_train_path='' # surprise의 train_data가 저장될 폴더의 경로(미리 생성해 두기!)
surprise_test_path='' # surprise의 test_data가 저장될 폴더의 경로


# 감정별로 사진을 각각 분리하여 train과 test 폴더에 넣어주기
data_split(angry, angry_test_cnt, angry_train_path, angry_test_path)
data_split(disgust, disgust_test_cnt, disgust_train_path, disgust_test_path)
data_split(fear, fear_test_cnt, fear_train_path, fear_test_path)
data_split(happy, happy_test_cnt, happy_train_path, happy_test_path)
data_split(neutral, neutral_test_cnt, neutral_train_path, neutral_test_path)
data_split(sad, sad_test_cnt, sad_train_path, sad_test_path)
data_split(surprise, surprise_test_cnt, surprise_train_path, surprise_test_path)
```
